**SHAPE CALCULATOR WITH MAGIC METHOD**

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # 2. __str__ method for readable output
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"

    # 3. Area calculation
    def area(self):
        return self.width * self.height

    # 4. Overloading '+' operator
    def __add__(self, other):
        # Adding dimensions to create a new Rectangle object
        new_width = self.width + other.width
        new_height = self.height + other.height
        return Rectangle(new_width, new_height)

    # 5. Overloading '==' operator to compare areas
    def __eq__(self, other):
        if isinstance(other, Rectangle):
            return self.area() == other.area()
        return False

# 6. Creating instances
rect1 = Rectangle(4, 5)
rect2 = Rectangle(2, 3)
# rect4 for equality comparison (same area as rect2)
rect4 = Rectangle(3, 2) 

# 7. Printing rectangles and areas
print(f"Rectangle 1: {rect1}")
print(f"Rectangle 2: {rect2}")
print(f"Area of rect1: {rect1.area()}")

# 8. Creating rect3 using '+' operator
rect3 = rect1 + rect2
print(f"rect3 (rect1 + rect2): {rect3}")

# 9. Comparing rectangles
print(f"Are rect1 and rect2 equal? {rect1 == rect2}")
print(f"Are rect2 and rect4 equal? {rect2 == rect4}")
```
