**NumPy Array Math Wizard**

```python
import numpy as np

#Create the initial arrays
# arr1 is a 2D array (2x3), arr2 is a 1D array (3,)
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([10, 20, 30])

#Perform Arithmetic Operations

addition       = arr1 + arr2
subtraction    = arr1 - arr2
multiplication = arr1 * arr2
division       = arr1 / arr2
```

```python
arr3 = np.array([1, 4, 9, 16, 25])

sqrt_arr = np.sqrt(arr3)  # Square root: [1. 2. 3. 4. 5.]
exp_arr  = np.exp(arr3)   # Exponential (e^x)
sin_arr  = np.sin(arr3)   # Sine of each element
```

```python
arr4 = np.array([[5, 12, 3], [8, 2, 7]])

# Total sum of every element in the array
total_sum = np.sum(arr4) 

# Max in each column (moving vertically across rows)
max_col = np.max(arr4, axis=0) 

# Min in each row (moving horizontally across columns)
min_row = np.min(arr4, axis=1)
```

```python
# Convert arr1 from int to float
arr1_float = arr1.astype(float)

print(f"Original Dtype: {arr1.dtype}")
print(f"New Dtype: {arr1_float.dtype}")
```
