```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def display_info(self):                      #method display
        print(f"name: {self.name}")
        print(f"age:{self.age}")

#derived class 
class Student(Person):
    def __init__(self, name, age, student_id,major):
        #super to call parent constructor
        super().__init__(name, age)

        #specific attributes
        self.student_id = student_id
        self.major = major


    #overriding
    def display_info(self):
        super().display_info()
        print(f"Student ID:{self.student_id}")
        print(f"Major:{self.major}")


#derived class 
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major,thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic

    def display_info(self):
        super().display_info()
        print(f"Thesis Topic:{self.thesis_topic}")

#creating object and call
print("Person")
p1 = Person("ali" , 25)
p1.display_info()

print("Student")
s1 = Student("ahmad",26,"s34567", "IT")
s1.display_info()

print("Graduate Student")
gs1 = GraduateStudent("ahmad", 27 , "s45677","physics", "science")
gs1.display_info()
    
        
```
